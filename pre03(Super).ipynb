```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA

features, targets = pre_a_df.iloc[:, :-1], pre_a_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()

pipe = Pipeline([('pca', PCA(n_components=3)), ('l_r', l_r) ])
pipe.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA

# 데이터 분할
X_train, X_test, y_train, y_test = \
    train_test_split(features, targets, test_size=0.2, random_state=124)

# PCA 및 Polynomial Regression 파이프라인 설정
pipe = Pipeline([
    ('pca', PCA(n_components=3)),  # PCA 차원 축소
    ('poly', PolynomialFeatures(degree=4)),
    ('regression', LinearRegression())  # 다항 회귀 모델
])

# 파이프라인 학습
pipe.fit(X_train, y_train)
```

### Cycle1-3. Others Regression Model

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error
from sklearn.model_selection import train_test_split

# Define a function to calculate evaluation metrics
def calculate_evaluation(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    if (y_true <= 0).any() or (y_pred <= 0).any():
        msle = np.nan
        rmsle = np.nan
    else:
        msle = mean_squared_log_error(y_true, y_pred)
        rmsle = np.sqrt(msle)

    return mse, rmse, msle, rmsle, r2

# Split data

features, targets = pre_a_df.iloc[:, :-1], pre_a_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# Define models with PCA and model itself as part of the pipeline
models = [
    ("Decision Tree", Pipeline([("scaler", StandardScaler()), ("pca", PCA()), ("dt", DecisionTreeRegressor(random_state=124))])),
    ("Random Forest", Pipeline([("scaler", StandardScaler()), ("pca", PCA()), ("rf", RandomForestRegressor(random_state=124, n_estimators=1000, n_jobs=-1))])),
    ("Gradient Boosting", Pipeline([("scaler", StandardScaler()), ("pca", PCA()), ("gb", GradientBoostingRegressor(random_state=124))])),
    ("XGBoost", Pipeline([("scaler", StandardScaler()), ("pca", PCA()), ("xgb", XGBRegressor())])),
    ("LightGBM", Pipeline([("scaler", StandardScaler()), ("pca", PCA()), ("lgb", LGBMRegressor(n_estimators=100))]))
]

# Iterate over models, fit, and evaluate
for name, model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    mse, rmse, msle, rmsle, r2 = calculate_evaluation(y_test, prediction)
    print('{} - MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'.format(name, mse, rmse, msle, rmsle, r2))
```

   (R2 점수가 가장 RandomForestRegressor이지만 데이터 양이 많고 점수차이가 제일 적게 나는 XGBRegressor를 사용해서 후를 진행했다.)
3. XGBRegressor를로 진행 했을 때, 훈련 데이터와 테스트 데이터의 R2 점수와 MSE 점수가 차이가 나는 것을 통해서 과적합임을 확인 했다.
4. 그 후 하이퍼 파마라미터를 조정해서 두 데이터들간의 R2 점수와 MSE 점수를 차이를 최대한으로 줄여보았다.
(결과적으론 과적합을 해소하지 못했다.)

따라서 해당 데이터 세트에 가장 적합한 모델은 XGBRegressor 이다.
